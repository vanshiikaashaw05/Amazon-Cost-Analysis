**Price Segmentation ‚Äì Low, Medium, High (NEW FEATURE)**

```python
def price_group(x):
    if x < 500: return 'Low Price'
    elif x < 2000: return 'Medium Price'
    else: return 'High Price'

ds['price_segment'] = ds['actual_price'].apply(price_group)

price_segment_count = ds['price_segment'].value_counts()

plt.figure(figsize=(8,6))
plt.pie(price_segment_count, labels=price_segment_count.index, autopct='%1.1f%%')
plt.title('Price Segment Distribution')
plt.show()
```
